```python
fname = "undirdir/1160295_101/jstor.tree"
```

```python
from collections import Counter, defaultdict
```

```python
c = Counter()
```

```python
size_total = Counter()
```

```python
with open(fname, 'r') as f:
    for line in f:
        if line[0] == "#":
            continue
            
        line = line.strip().split(' ')
        cl = line[0].split(":")
        depth = len(cl)
        
        belongs_to = ':'.join(cl[:-1])
        size_total[belongs_to] += 1
        c[depth] += 1
```

```python
c
```

```python
size_total.most_common(10)
```

```python
size_counts = Counter()
for count in size_total.itervalues():
    size_counts[count] += 1
```

```python
size_counts.most_common()
```
